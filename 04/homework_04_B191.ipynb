```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sklearn.preprocessing as pp

from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_excel('Residential-Building-Data-Set.xlsx', skiprows = [0])
df = df.rename(columns = {"V-9": "sales_prices", "V-10": "construction_costs"})
df
```

```python
X = df.drop(columns = ["sales_prices", "construction_costs"])
construction_costs = df.construction_costs.values
sales_prices = df.sales_prices.values
```

# Regrese

```python
def fit_model(model, X, Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
    m = model()
    m.fit(x_train, y_train)
    prediction = m.predict(x_test)
    
    model_name = model.__name__
    display(
        "Mean absolute error (" + model_name + "): " + str(mean_absolute_error(np.array(y_test), prediction))
    )
    return (m, y_test, prediction)

def plot_model(y_test, prediction, title):    
    comparison = pd.DataFrame({"test": y_test, "prediction": prediction})
    comparison.plot(figsize = (15, 7))
    plt.suptitle(title)
    plt.show()

def fit_and_plot_model(model, title, X, Y):
    m, y_test, prediction = fit_model(model, X, Y)
    plot_model(y_test, prediction, title)
```

```python
fit_and_plot_model(LinearRegression, "Chyba lineární regrese na sales_prices", X, sales_prices)
```

```python
fit_and_plot_model(Ridge, "Chyba hřebenové regrese na sales_prices", X, sales_prices)
```

```python
fit_and_plot_model(LinearRegression, "Chyba lineární regrese na construction_costs", X, construction_costs)
```

```python
fit_and_plot_model(Ridge, "Chyba hřebenové regrese na construction_costs", X, construction_costs)
```

# Normalizace dat
Náhled několika škálovacích metod a normalizace dat ukazuje, že vliv těchto předzpracovacích kroků je na výkon modelů zanedbatelný, alespoň v našem případě.

```python
for (n, Y) in [("sales_prices", sales_prices), ("construction_costs", construction_costs)]:
    print(" === " + n + " === ")
    for scaler in [pp.StandardScaler, pp.MaxAbsScaler, pp.RobustScaler, pp.MinMaxScaler, pp.Normalizer]:
        s = scaler()
        r = s.fit_transform(X, Y)
        display(s)
        for model in [LinearRegression, Ridge]:
            fit_model(model, X, Y)
        print()
        print()
```

# Ladění hyperparametrů hřebenové regrese
Využijeme `GridSearchCV` pro nalezení optimální hodnoty $\alpha$ hřebenové regrese. Výsledek předčil normalizaci i škálovací metody v předchozím bodě.

```python
def tune_ridge_regression(Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
    regressor = GridSearchCV(Ridge(), {
        'alpha': [pow(10, x) for x in range(-5, 3)],
    })
    regressor.fit(x_train, y_train)
    display(regressor.best_params_)
    prediction = regressor.predict(x_test)
    display(mean_absolute_error(np.array(y_test), prediction))
    return regressor

tune_ridge_regression(sales_prices)
```

# Náhodné lesy
`GridSearchCV` využijeme i pro náhodný les. Jak je vidět, ani relativně velký počet rozhodovacích stromů různé hloubky nedosahuje výsledků regresních přístupů výše.

```python
def try_forests(Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

    hyperparameters = {
        "n_estimators": range(10, 150, 15),
        "max_depth": [1, 2, 3, 4, 6, 8, 10],
    }

    grid = GridSearchCV(RandomForestRegressor(), hyperparameters, n_jobs = -1)
    grid.fit(x_train, y_train)
    display(mean_absolute_error(y_test, grid.best_estimator_.predict(x_test)))

try_forests(sales_prices)
```
